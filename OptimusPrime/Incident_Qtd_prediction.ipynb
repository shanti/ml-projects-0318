```python
# imports
%matplotlib inline

import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import seaborn as sns #seaborn produces pretty nice graphs!
from scipy import stats
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from scipy.stats.stats import pearsonr
from statsmodels.graphics.regressionplots import abline_plot
```

```python
events = pd.read_csv(r"\\ausemsmttp.aus.amer.dell.com\mlproject\EventsV4.csv")
events
```

```python
# TODO: select only "SalePrice", "Bedrooms", "Bathrooms", "SqFtTotLiving", "SqFtLot"
events_small = events[["EventTypeId","ImpactTypeId", "IncsCount"]]
events_small
```

```python
#TESTING DATA PREPARATION

## TODO : input cols : gre, gpa, rank
fv = events_small[['EventTypeId', 'ImpactTypeId']].values
labels = events_small['IncsCount'].values


X_train, X_test, Y_train, Y_test = train_test_split(events_small[['EventTypeId', 'ImpactTypeId']], labels, test_size=0.2)
print("training set = " , len(X_train))
print("testing set = " , len(Y_test))
```

```python
clf = svm.SVC(kernel='linear')

t1 = time.perf_counter()
clf.fit(X_train, Y_train)
t2 = time.perf_counter()


print("trained on {:,} records in {:,.2f} ms".\
      format(len(X_train), (t2-t1)*1000))

# Print the coefficients and intercept for linearsSVC
print ("inputs : EventTypeId, ImpactTypeId, IncsCount")
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
print("Accuracy = " + str(accuracy_score(Y_test, predictions)))
```
